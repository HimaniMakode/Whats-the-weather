```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

```python

# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name

    
    # If the city is unique, then add it to a our cities list
    if city not in cities:

        cities.append(city)

# Print the city count 
len(cities)
```

```python
# OpenWeatherMap API Key
# api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key
url
```

```python
# Create empty lists to append the API data into lists 
city_name = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []

# Create row counter and set it to one
row_counter = 1

#print statement
print("Beginning Data Retrieval")
print("---------------------------------------------------")

#Looping through the cities in the city list 
for city in cities:
    
    #Try and except function to append the calls where the value is found and skip to next when not found
    try:
        
        response = requests.get(f"{url}&q={city}").json() 
        city_name.append(response["name"])
        cloudiness.append(response["clouds"]["all"])
        country.append(response["sys"]["country"])
        date.append(response["dt"])
        humidity.append(response["main"]["humidity"])
        max_temp.append(response["main"]["temp_max"])
        lat.append(response["coord"]["lat"])
        lng.append(response["coord"]["lon"])
        wind_speed.append(response["wind"]["speed"])
        city_record = response["name"]
        print(f"Processing Record {row_counter} | {city_record}")
        print(f"{url}&q={city}")
    
        #Increment row_counter
        row_counter = row_counter + 1
        
        #Pause program execution for 1 sec
        time.sleep(1.00)
        
    except:
        print("City not found. Skipping...")
    continue
```

```python
#Create a dictionary

weather_dict = {
    
    "City": city_name,
    "Cloudiness": cloudiness,
    "Country": country,
    "Date": date,
    "Humidity": humidity,
    "Lat": lat,
    "Lng": lng,
    "Max Temp": max_temp,
    "Wind Speed": wind_speed,
}

#Convert into DataFrame
weather_df = pd.DataFrame(weather_dict)

weather_df.head()

```

```python
#To display the count of the weather data

weather_df.count()
```

```python
#Save the dataframe to csv

weather_df.to_csv("output_data/weather_data.csv")
```

```python
# scatter plot for Latitude vs. Temperature

weather_df.plot(kind= "scatter", x= "Lat", y= "Max Temp", marker = "o", facecolor = "blue",
                                edgecolors = "black", s= 120, figsize=(20,15))

# Create a title, x label, and y label for our chart

plt.title("Latitude vs. Temperature", fontsize= 30)
plt.xlabel("Latitude", fontsize= 25)
plt.ylabel("Max Temperature(F)", fontsize= 25)
plt.grid()

#Save the image
plt.savefig("output_images/Latitude_vs_Temperature.png")

plt.show()
```
